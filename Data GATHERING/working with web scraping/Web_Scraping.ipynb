```python
final=pd.DataFrame()
for j in range(1,28):
    headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5)AppleWebKit/605.1.15 (KHTML, like Gecko)Version/12.1.1 Safari/605.1.15'}
    webpage = requests.get('https://www.trustpilot.com/categories/fitness_and_nutrition_service?page={}'.format(j), headers= headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_="paper_paper__1PY90 paper_outline__lwsUX card_card__lQWDv card_noPadding__D8PcU styles_wrapper__2JOo2")
    name=[]
    rating=[]
    reviews=[]
    location=[]
    fitness_type=[]
    for i in company:
        try:
            name.append(i.find('p',class_="typography_heading-xs__jSwUz typography_appearance-default__AAY17 styles_displayName__GOhL2").text.strip())
        except:
            name.append(np.nan)
        try:
            rating.append(i.find('span',class_="typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_trustScore__8emxJ").text.strip())
        except:
            rating.append(np.nan)
        try:
            reviews.append(i.find('p',class_='typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_ratingText__yQ5S7').text.strip())
        except:
            reviews.append(np.nan)
        try:
            location.append(i.find('span',class_="typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_metadataItem__Qn_Q2 styles_location__ILZb0").text.strip())
        except:
            location.append(np.nan)
        try:
            fitness_type.append(i.find('div',class_='styles_wrapper___E6__ styles_categoriesLabels__FiWQ4 styles_desktop__U5iWw').text.strip())
        except:
            fitness_type.append(np.nan)
            
    df=pd.DataFrame({'name':name,
                     'Rating':rating,
                     'Reviews':reviews,
                     'Location':location,
                     'Fitness':fitness_type,
                    })
    final=final.append(df,ignore_index=True)
    
```
